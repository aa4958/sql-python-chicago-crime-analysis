Query the database system catalog to retrieve column metadata¶
The SCHOOLS table contains a large number of columns. How many columns does this table have?

```python
pd.read_sql("SELECT count(name) FROM PRAGMA_TABLE_INFO('Chicago')",con)
```

```python
pd.read_sql("select name,type, length(type) FROM PRAGMA_TABLE_INFO('CHICAGO')",con)
```

```python
pd.read_sql("""
SELECT count(*) AS cnt
FROM Chicago
WHERE "Elementary, Middle, or High School" = "ES";
""", con)
```

```python
pd.read_sql("""
SELECT "Name of School", "Safety Score"
FROM Chicago
WHERE "Safety Score" = (
    SELECT MAX("Safety Score") FROM Chicago
);
""", con)
```

```python
pd.read_sql("""
SELECT
    "Name of School",
    "Average Student Attendance"
FROM Chicago
WHERE "Average Student Attendance" IS NOT NULL
ORDER BY "Average Student Attendance" DESC
LIMIT 10;
""", con)
```

```python
pd.read_sql("""
SELECT
    "Name of School",
    "Average Student Attendance"
FROM Chicago
WHERE "Average Student Attendance" IS NOT NULL
ORDER BY "Average Student Attendance" 
Limit 5;


""",con)
```

```python
pd.read_sql("""
SELECT "Name of School", REPLACE("Average Student Attendance", '%', '') as "Average Student Attendance"
     from Chicago 
     order by "Average Student Attendance "
     LIMIT 5


""",con)
```
