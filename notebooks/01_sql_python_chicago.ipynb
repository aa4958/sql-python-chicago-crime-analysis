# SQL + Python Chicago Database Analysis

This notebook loads CSV datasets into a local SQLite database (`outputs/Final.db`) and runs SQL queries with results displayed in pandas.


```python
import pandas as pd
import sqlite3 as sq

# connect to database (will create if doesn't exist)
con = sq.connect('outputs/Final.db')

# load CSVs
census = pd.read_csv('data/ChicagoCensusData.csv')
crime = pd.read_csv('data/ChicagoCrimeData.csv')
schools = pd.read_csv('data/ChicagoPublicSchools.csv')

# write tables
census.to_sql('CENSUS', con, if_exists='replace', index=False)
crime.to_sql('CRIME', con, if_exists='replace', index=False)
schools.to_sql('SCHOOLS', con, if_exists='replace', index=False)

# example query: total crime incidents
pd.read_sql('SELECT count(*) as incident_count FROM CRIME;', con)
```

```python
# more queries can be added here
pd.read_sql("SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS incident_count FROM CRIME GROUP BY COMMUNITY_AREA_NUMBER ORDER BY incident_count DESC LIMIT 5;", con)
```
